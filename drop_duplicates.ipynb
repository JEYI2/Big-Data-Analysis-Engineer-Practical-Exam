## f1의 결측치를 채운 후 age 컬럼의 중복 제거 전과 후의 'f1' 중앙값 차이를 구하시오 
### - 결측치는 'f1' 데이터에서 큰 값 순으로 정렬했을 때 10번째에 위치한 값으로 채운다.
### - 중복 데이터 발생시 뒤에 나오는 데이터를 삭제함
### - 최종 결과값은 절대값으로 출력
(중복 제거 기준: 제공된 데이터 순서대로 중복 데이터 발생시 뒤에 나오는 데이터를 삭제함)


- 데이터셋 : basic1.csv
- 오른쪽 상단 copy&edit 클릭 -> 예상문제 풀이 시작

```python
import pandas as pd
```

```python
df = pd.read_csv('/kaggle/input/bigdatacertificationkr/basic1.csv')
df
```

```python
#상위 10번째 f1값
df_sort = df.sort_values(by ='f1' ,ascending =False)
f1_10 = df_sort['f1'][9:10].values[0]
print('f1 상위 10번째 값:', f1_10)
```

```python
# 누락값 채우기
df['f1'] = df['f1'].fillna(f1_10)
before_median = df['f1'].median()
# age 중복 제거 전 f1 중앙값 
print('age 중복 제거 전 f1 중앙값:',before_median)
```

```python
# age 중복 제거 후 f1 중앙값 
df_duplicates = df.drop_duplicates(subset = 'age', keep = 'first')
after_median = df_duplicates['f1'].median()
print('age 중복 제거 후 f1 중앙값:',after_median)
```
