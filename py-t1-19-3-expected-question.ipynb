### 단 Events컬럼이 '1'인경우 80%의 Sales값만 반영함

```python
# 라이브러리 불러오기
import pandas as pd

pd.set_option('display.max_column', 10)
pd.set_option('display.max_row',j10)
df = pd.read_csv('/kaggle/input/bigdatacertificationkr/basic2.csv')
df
```

```python
df = pd.read_csv('/kaggle/input/bigdatacertificationkr/basic2.csv')
df.loc[df['Events'] == 1, 'Sales'] *= 0.8
df
```

```python
df['Year'] = pd.to_datetime(df['Date']).dt.year
df['Month'] = pd.to_datetime(df['Date']).dt.month
```

```python
df_2022= df[df['Year'] == 2022].groupby('Month')['Sales'].sum().max()
df_2023 = df[df['Year'] == 2023].groupby('Month')['Sales'].sum().max()
print(df_2022)
print(df_2023)
abs(round(df_2022-df_2023))
```
